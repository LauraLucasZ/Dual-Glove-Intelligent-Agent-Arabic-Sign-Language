## 6b. Promote comparison winners to chosen (move)

Files moved to `chosen/` are removed from `comparison/` (no duplicates).


```python
import shutil



COPY_CHOSEN_LOG = []

REMOVE_FROM_COMPARISON_LOG = []



print("=== Promote comparison winners -> chosen (move) ===")

for chosen_stem, fallbacks in COPY_TO_CHOSEN:

    sources = [chosen_stem, *fallbacks]

    src_path = None

    used_stem = None

    for stem in sources:

        candidate = DIR_COMPARISON / f"{stem}.mp3"

        if candidate.is_file():

            src_path = candidate

            used_stem = stem

            break



    dest = DIR_CHOSEN / f"{chosen_stem}.mp3"



    if src_path is not None:

        if dest.is_file():

            removed_name = src_path.name

            src_path.unlink()

            print(f"chosen: {chosen_stem}.mp3 (kept; removed duplicate {removed_name} from comparison)")

            print(f"output MP3 path: {dest.resolve()}")

        else:

            shutil.move(str(src_path), str(dest))

            action = "moved" if used_stem == chosen_stem else f"moved (from {used_stem})"

            print(f"chosen: {chosen_stem}.mp3 <- {used_stem}.mp3 ({action})")

            print(f"output MP3 path: {dest.resolve()}")

        COPY_CHOSEN_LOG.append({"chosen_stem": chosen_stem, "source": used_stem, "path": dest.resolve()})

    elif dest.is_file():

        print(f"chosen: {chosen_stem}.mp3 (already in chosen)")

        print(f"output MP3 path: {dest.resolve()}")

        COPY_CHOSEN_LOG.append({"chosen_stem": chosen_stem, "source": None, "path": dest.resolve()})

    else:

        print(f"MISSING: none of {sources} found in comparison or chosen")

        print("---")

        continue



    for stem in sources:

        comp_mp3 = DIR_COMPARISON / f"{stem}.mp3"

        if comp_mp3.is_file():

            comp_mp3.unlink()

            REMOVE_FROM_COMPARISON_LOG.append(stem)

            print(f"removed from comparison: {stem}.mp3")

    print("---")



print(f"Moved/promoted to chosen: {len(COPY_CHOSEN_LOG)}/{len(COPY_TO_CHOSEN)}")

print(f"Removed from comparison: {len(REMOVE_FROM_COMPARISON_LOG)} file(s)")

if REMOVE_FROM_COMPARISON_LOG:

    print("  stems:", ", ".join(REMOVE_FROM_COMPARISON_LOG))



```

## 6c. Cut trailing audio (`comp_qaf_ق_qaf_split`)

Remove ~last 40s of noise when present; otherwise trim trailing silence. Saves to `comparison/cut/` without overwriting the original.


```python
import subprocess

import sys



def _ensure_pkg(pkg: str, import_name: str | None = None):

    name = import_name or pkg

    try:

        return __import__(name, fromlist=[""])

    except ImportError:

        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", pkg])

        return __import__(name, fromlist=[""])



_ensure_pkg("mutagen")

_ensure_pkg("pydub")

from mutagen.mp3 import MP3

from pydub import AudioSegment

from pydub.silence import detect_nonsilent

from IPython.display import Audio as IPA, display



TRIM_TAIL_SEC = 40.0

QAF_SPLIT_STEM = "comp_qaf_ق_qaf_split"

QAF_SPLIT_ORIG = DIR_COMPARISON / f"{QAF_SPLIT_STEM}.mp3"

QAF_SPLIT_CUT = DIR_COMPARISON_CUT / f"{QAF_SPLIT_STEM}_cut.mp3"



DIR_COMPARISON_CUT.mkdir(parents=True, exist_ok=True)



if not QAF_SPLIT_ORIG.is_file():

    raise FileNotFoundError(f"Missing original: {QAF_SPLIT_ORIG}")



audio = AudioSegment.from_mp3(QAF_SPLIT_ORIG)

orig_duration_sec = len(audio) / 1000.0



if orig_duration_sec > TRIM_TAIL_SEC:

    keep_ms = int((orig_duration_sec - TRIM_TAIL_SEC) * 1000)

    cut_audio = audio[:keep_ms]

    trim_note = f"removed last {TRIM_TAIL_SEC:.0f}s"

else:

    nonsilent = detect_nonsilent(

        audio, min_silence_len=80, silence_thresh=audio.dBFS - 14, seek_step=10

    )

    if nonsilent:

        end_ms = min(len(audio), nonsilent[-1][1] + 120)

        cut_audio = audio[:end_ms]

        trim_note = "trimmed trailing silence/noise (file shorter than 40s)"

    else:

        keep_ms = max(300, int(len(audio) * 0.75))

        cut_audio = audio[:keep_ms]

        trim_note = "kept first 75% (short clip, no silence markers)"



cut_audio.export(QAF_SPLIT_CUT, format="mp3")

cut_duration_sec = len(cut_audio) / 1000.0



print("=== Cut comp_qaf_ق_qaf_split ===")

print(f"original: {QAF_SPLIT_ORIG.resolve()}")

print(f"original duration: {orig_duration_sec:.3f}s")

print(f"cut duration: {cut_duration_sec:.3f}s ({trim_note})")

print(f"saved path: {QAF_SPLIT_CUT.resolve()}")



print("\n--- Playback: original ---")

display(IPA(filename=str(QAF_SPLIT_ORIG)))

print("--- Playback: cut ---")

display(IPA(filename=str(QAF_SPLIT_CUT)))

```

## 4. Select Edge TTS voice

```python
import asyncio

import edge_tts
import nest_asyncio

nest_asyncio.apply()

PREFERRED_VOICES = ["ar-EG-SalmaNeural", "ar-EG-ShakirNeural"]
EDGE_VOICE = None
EDGE_TTS_LOADED = False
_ALL_VOICES = None

async def _list_voices():
    global _ALL_VOICES
    if _ALL_VOICES is None:
        _ALL_VOICES = await edge_tts.list_voices()
    return _ALL_VOICES

async def select_edge_voice() -> str:
    voices = await _list_voices()
    by_short = {v["ShortName"]: v for v in voices}

    for pref in PREFERRED_VOICES:
        if pref in by_short:
            return pref

    ar_names = sorted(
        v["ShortName"]
        for v in voices
        if str(v.get("Locale", "")).lower().startswith("ar")
    )
    print("Preferred voices unavailable. Available Arabic voices:")
    for name in ar_names:
        print(" ", name)

    eg = [
        v["ShortName"]
        for v in voices
        if "EG" in v.get("Locale", "") or "-EG-" in v.get("ShortName", "")
    ]
    if eg:
        return eg[0]
    if ar_names:
        return ar_names[0]
    raise RuntimeError("No Arabic Edge TTS voice found")

EDGE_VOICE = asyncio.get_event_loop().run_until_complete(select_edge_voice())
EDGE_TTS_LOADED = True
print(f"Selected Edge TTS voice: {EDGE_VOICE}")
```

## 5. Edge TTS synthesis (MP3 only)


```python
import re

import time

from typing import Optional



from IPython.display import Audio as IPA, display



CHOSEN_LOG = []

COMPARISON_LOG = []

PLAYBACK_SHOWN = False



def format_bytes(nb: int) -> str:

    if nb < 1024:

        return f"{nb} B"

    if nb < 1024**2:

        return f"{nb / 1024:.2f} KB"

    return f"{nb / 1024 / 1024:.2f} MB"



async def synth_edge_mp3(text: str, mp3_path: Path) -> float:

    mp3_path.parent.mkdir(parents=True, exist_ok=True)

    t0 = time.perf_counter()

    comm = edge_tts.Communicate(text, EDGE_VOICE)

    await comm.save(str(mp3_path))

    return time.perf_counter() - t0



async def generate_mp3(

    letter: str,

    pron: str,

    mp3_path: Path,

    show_playback: bool = False,

    target_sound: str | None = None,

) -> dict | None:

    global PLAYBACK_SHOWN

    tts_text = with_period(pron)

    try:

        infer_s = await synth_edge_mp3(tts_text, mp3_path)

    except Exception as exc:

        if mp3_path.is_file() and mp3_path.stat().st_size == 0:

            mp3_path.unlink()

        print(f"letter: {letter}")

        if target_sound:

            print(f"target sound: {target_sound}")

        print(f"variant text sent to Edge TTS: {tts_text}")

        print(f"FAILED (no audio): {exc}")

        print("---")

        return None

    if not mp3_path.is_file():

        print(f"letter: {letter}")

        print(f"variant text sent to Edge TTS: {tts_text}")

        print("FAILED:", mp3_path)

        print("---")

        return None

    entry = {

        "letter": letter,

        "pron": pron,

        "tts_text": tts_text,

        "mp3_path": mp3_path.resolve(),

        "infer_s": infer_s,

        "bytes": mp3_path.stat().st_size,

    }

    print(f"letter: {letter}")

    if target_sound:

        print(f"target sound: {target_sound}")

    print(f"variant text sent to Edge TTS: {tts_text}")

    print(f"output MP3 path: {entry['mp3_path']}")

    print(f"inference: {infer_s:.3f}s | {format_bytes(entry['bytes'])}")

    print("---")

    if show_playback:

        display(IPA(filename=str(mp3_path)))

        PLAYBACK_SHOWN = True

    return entry



def run_async(coro):

    try:

        asyncio.get_running_loop()

    except RuntimeError:

        return asyncio.run(coro)

    return asyncio.get_event_loop().run_until_complete(coro)

```

## 6. Generate chosen mappings (MP3)


```python
import shutil

CHOSEN_LOG.clear()
CHOSEN_OK = 0

print("=== Chosen mappings ===")
for stem, letter, pron in CHOSEN_MAPPINGS:
    mp3_path = DIR_CHOSEN / f"{stem}.mp3"
    comp_path = DIR_COMPARISON / f"{stem}.mp3"
    tts_text = with_period(pron)

    if mp3_path.is_file() and stem in CHOSEN_SKIP_REGEN:
        print(f"letter: {letter}")
        print(f"variant text sent to Edge TTS: {tts_text}")
        print(f"output MP3 path: {mp3_path.resolve()} (kept, skipped regen)")
        print("---")
        CHOSEN_LOG.append({
            "stem": stem, "letter": letter, "pron": pron, "tts_text": tts_text,
            "mp3_path": mp3_path.resolve(), "infer_s": 0.0, "bytes": mp3_path.stat().st_size,
        })
        CHOSEN_OK += 1
        continue

    if stem in CHOSEN_SKIP_REGEN and comp_path.is_file() and not mp3_path.is_file():
        shutil.copy2(comp_path, mp3_path)
        print(f"letter: {letter}")
        print(f"variant text sent to Edge TTS: {tts_text}")
        print(f"output MP3 path: {mp3_path.resolve()} (copied from comparison)")
        print("---")
        CHOSEN_LOG.append({
            "stem": stem, "letter": letter, "pron": pron, "tts_text": tts_text,
            "mp3_path": mp3_path.resolve(), "infer_s": 0.0, "bytes": mp3_path.stat().st_size,
        })
        CHOSEN_OK += 1
        continue

    if mp3_path.is_file():
        print(f"letter: {letter}")
        print(f"variant text sent to Edge TTS: {tts_text}")
        print(f"output MP3 path: {mp3_path.resolve()} (kept)")
        print("---")
        CHOSEN_LOG.append({
            "stem": stem, "letter": letter, "pron": pron, "tts_text": tts_text,
            "mp3_path": mp3_path.resolve(), "infer_s": 0.0, "bytes": mp3_path.stat().st_size,
        })
        CHOSEN_OK += 1
        continue

    entry = run_async(generate_mp3(letter, pron, mp3_path, show_playback=False))
    if entry:
        entry["stem"] = stem
        CHOSEN_LOG.append(entry)
        CHOSEN_OK += 1

print(f"Chosen: {CHOSEN_OK}/{len(CHOSEN_MAPPINGS)} MP3 files")
```
