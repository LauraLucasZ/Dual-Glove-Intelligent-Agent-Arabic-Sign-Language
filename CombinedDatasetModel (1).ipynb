```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GroupShuffleSplit  
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import joblib
```

```python
file_path = r"C:\Users\moham\Downloads\numbers_letters_dataset_sorted.csv"

df = pd.read_csv(file_path)

print(df.head())
```

```python
print("Dataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("Unique Labels:")
print(df['label'].unique())

print("\nLabel Counts:")
print(df['label'].value_counts())
```

```python
X = df[[
    'flex1', 'flex2', 'flex3', 'flex4', 'flex5',
    'accX', 'accY', 'accZ',
    'gyroX', 'gyroY', 'gyroZ'
]]

y = df['label']

print("Features Shape:", X.shape)
print("Labels Shape:", y.shape)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Features scaled successfully")
print(X_scaled[:5])
```

```python
noise = np.random.normal(
    loc=0,
    scale=0.03,
    size=X_scaled.shape
)

X_scaled = X_scaled + noise

print("Noise augmentation applied")
```

```python
frames_per_session = 40


df['session_id'] = np.arange(len(df)) // frames_per_session

print(df[['label', 'session_id']].head(50))
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y_encoded = label_encoder.fit_transform(y)

print("Labels encoded successfully")
print(y_encoded[:10])
```

```python
groups = df['session_id']

gss = GroupShuffleSplit(
    n_splits=1,
    test_size=0.2,
    random_state=42
)

train_idx, test_idx = next(
    gss.split(
        X_scaled,
        y_encoded,
        groups
    )
)

X_train = X_scaled[train_idx]
X_test = X_scaled[test_idx]

y_train = y_encoded[train_idx]
y_test = y_encoded[test_idx]

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier

model = HistGradientBoostingClassifier(
    learning_rate=0.05,
    max_iter=200,
    max_depth=6,
    min_samples_leaf=20,
    l2_regularization=1.0,
    random_state=42
)

model.fit(X_train, y_train)

print("Model Training Complete")
```

```python
y_pred = model.predict(X_test)
```

```python
labels_in_test = np.unique(y_test)

report = classification_report(
    y_test,
    y_pred,
    labels=labels_in_test,
    target_names=label_encoder.inverse_transform(labels_in_test)
)

print(report)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
joblib.dump(model, 'CombinedGesture_model.pkl')
joblib.dump(scaler, 'Modelscaler.pkl')
joblib.dump(label_encoder, 'label_encoder.pkl')

print("Model Saved Successfully")
```
